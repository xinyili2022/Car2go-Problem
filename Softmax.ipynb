# Upload and Preprocess the Dataset

```python
urls = [
    "http://pin13.net/car2go/data/2012-06.txt.gz",
    "http://pin13.net/car2go/data/2012-07.txt.gz",
    "http://pin13.net/car2go/data/2012-08.txt.gz",
    "http://pin13.net/car2go/data/2012-09.txt.gz",
    "http://pin13.net/car2go/data/2012-10.txt.gz",
    "http://pin13.net/car2go/data/2012-12.txt.gz",
    "http://pin13.net/car2go/data/2013-01.txt.gz",
    "http://pin13.net/car2go/data/2013-02.txt.gz",
    "http://pin13.net/car2go/data/2013-03.txt.gz",
    "http://pin13.net/car2go/data/2013-04.txt.gz",
    "http://pin13.net/car2go/data/2013-05.txt.gz",
    "http://pin13.net/car2go/data/2013-06.txt.gz",
    "http://pin13.net/car2go/data/2013-07.txt.gz",
    "http://pin13.net/car2go/data/2013-08.txt.gz",
    "http://pin13.net/car2go/data/2013-09.txt.gz",
    "http://pin13.net/car2go/data/2013-10.txt.gz",
    "http://pin13.net/car2go/data/2013-11.txt.gz"]

for url in urls:
    filename = url.split('/')[-1]

    response = requests.get(url, stream=True)
    response.raise_for_status()

    with open(filename, 'wb') as file:
        for chunk in response.iter_content(chunk_size=8192):
            file.write(chunk)

    print(f"File {filename} downloaded successfully!")
```

```python
#combine the dataset
filenames = ["2012-06.txt.gz"]

#filenames = [
    #"2012-06.txt.gz",
    #"2012-07.txt.gz"
   # ]

Data = pd.concat([pd.read_csv(f) for f in filenames],ignore_index=True)
```

Found there are leading spaces in the names of the columns and the values, then we get rid of the spaces

```python
Data.columns = [col.strip() for col in Data.columns]  #get rid of the leading space in the name of coulmns
```

```python
Data = Data.applymap(lambda x: x.strip() if isinstance(x, str) else x) #get rid of the leading space in the values
```

Separate the date and time

```python
#sperate the Real date and time
Data['time'] = pd.to_datetime(Data['date']).dt.time
Data['date'] = pd.to_datetime(Data['date']).dt.date
```

```python
#adjust the order of columns
desired_order = ['license', 'date', 'time', 'latitude', 'longitude', 'fuel', 'interior', 'exterior',]
Data = Data[desired_order]
```

Change the results in interior and exterior to catergorical number

```python
# Convert true and false to 1 and 0, notice that is string not boolean
for col in ['interior','exterior']:
    Data[col] = Data[col].replace({'true': 1, 'false': 0})
```

Sort the records

```python
Data.sort_values(by=['license', 'date', 'time'], inplace=True) #sort the records as license, date and time
```

Set the difference

```python
# Calculate the difference between consecutive rows for 'latitude', 'longitude', and 'fuel'
Data['latitude_diff'] = Data['latitude'].diff()
Data['longitude_diff'] = Data['longitude'].diff()
Data['fuel_diff'] = Data['fuel'].diff()

filter = (Data['latitude_diff'] != 0) & (Data['longitude_diff'] != 0) & (Data['fuel_diff'] <= 0)
```

```python
print(len(Data[filter])) #check the number of pick-up
```

Set the two consecutive rows

```python
collection_set = filter | filter.shift(-1) #combine the needed rows

Data_collection= Data[collection_set]
Data_collection = Data_collection.drop(['latitude_diff', 'longitude_diff', 'fuel_diff'], axis=1)
```

```python
print(Data_collection.head(30))
```

Add the another column to distinguish, setting the former one as 1, the latter one as 2

```python
Data_collection_new= Data[collection_set]
Data_collection_new['group'] = None

#setting original filter as 2
Data_collection_new.loc[filter, 'group'] = 2
#shift back one row, setting as 1
shift_filter = (Data_collection_new['group'].shift(-1) == 2) & collection_set
Data_collection_new.loc[shift_filter, 'group'] = 1

# Drop the diff columns if no longer needed
Data_collection_new = Data_collection_new.drop(['latitude_diff', 'longitude_diff', 'fuel_diff'], axis=1)

# Display the head of the DataFrame
print(Data_collection_new.head(35))
```

Check the duration of trips

```python
# Get the indices of the filtered rows
filter_index = Data.index[filter]

# Create a boolean series
duration_check = pd.Series(False, index=Data.index)

for index in filter_index: #for loop
    if index + 1 < len(Data): # within the range
           if Data.at[index, 'longitude'] != Data.at[index + 1, 'longitude']:  # Check whether the difference in longitude is non-zero, only needed to do once, the results of latitude must be same
               duration_check.at[index] = True

check_rows = Data[duration_check]
```
